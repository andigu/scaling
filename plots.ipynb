```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv("~/scratch/scaling/curriculum_learning/time-chunk-lowp-curriculum/logs/metrics.csv")[-10:].mean()
df2 = pd.read_csv("~/scale/src/samples_d9_highp.csv")
delta = np.array([df[f'inacc_{i}'] for i in range(19)])/df2['error_rate']
plt.plot(delta)
plt.xlabel('Number of rounds')
plt.ylabel('Final inaccuracy / MWPM inaccuracy')
```

```python
df = pd.read_csv("~/scratch/scaling/curriculum_learning/time-chunk-lowp-curriculum/logs/metrics.csv")
df = df[10:]
df2 = pd.read_csv("~/scale/src/samples_d9_highp.csv")
for i in range(19):
    std_error_rate = df2['std_error_rate'].iloc[i]
    error_rate = df2['error_rate'].iloc[i]
    print(std_error_rate)
    plt.fill_between(df['step'], df[f'inacc_{i}']/(error_rate - std_error_rate), df[f'inacc_{i}']/(error_rate + std_error_rate), alpha=0.2)
plt.axhline(1.0)
plt.ylim(None, 1.1)
```
