```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import scipy
df = pd.read_csv("/n/home07/andigu/scale/src/logs/conv3d/version_60/metrics.csv")
fig, axes = plt.subplots(1, 2, figsize=(8, 4), sharex=True)
data_amt = np.logspace(np.log10(df['step']*512).min(), 7.5)
cmap = plt.get_cmap('plasma')
nrounds = 11
for i in range(nrounds):
    x, y = df['step']*1536, df[f'inacc_{i}']
    x, y = x[y>0], y[y>0]
    if len(y)>0:

        axes[0].loglog(x, y, label=f'Round {i}', c=cmap(i/13))
        interp = scipy.interpolate.make_smoothing_spline(np.log(x), np.log(y), lam=0.1)
        # axes[0].loglog(x, np.exp(interp(np.log(x))), c=f'C{i}')
        axes[1].plot(x, -interp.derivative()(np.log(x)), c=cmap(i/nrounds))
        x, y = x[len(x)//2:], y[len(y)//2:]
        m,b = np.polyfit(np.log(x), np.log(y), deg=1)
        print(m)
        axes[0].plot(data_amt, np.exp(b) * data_amt**m, linestyle='--', alpha=0.5, c=cmap(i/nrounds), lw=1)
axes[0].axhline(0.0711, c='black')
axes[0].grid()
axes[0].grid(which='minor', linestyle=':', linewidth='0.5', color='gray')
axes[0].set_xlim(3e5, 5e7)
axes[0].set_ylim(1e-3, 0.4)
axes[1].grid()
```

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import scipy
def plot_perf(vnum, fig, axes, nrounds, baseline=0.0588, s=0.05):
    df = pd.read_csv(f"/n/home07/andigu/scale/src/logs/conv3d/version_{vnum}/metrics.csv")
    data_amt = np.logspace(np.log10(df['step']*512).min(), 7.5)
    cmap = plt.get_cmap('cividis')
    for i in range(nrounds):
        x, y = df['step']*512, df[f'inacc_{i}']
        x, y = x[y>0], y[y>0]
        if len(y)>0:

            
            # interp = scipy.interpolate.make_smoothing_spline(np.log(x), np.log(y), lam=0.1)
            interp = scipy.interpolate.make_splrep(np.log(x)[20:], np.log(y)[20:], k=1, s=s)
            axes[0].loglog(data_amt, np.exp(interp(np.log(data_amt))), c=cmap(i/nrounds), ls='--', alpha=0.3, lw=4)
            axes[0].loglog(x, y, label=f'Round {i}', c=cmap(i/13))
            axes[1].plot(x, -interp.derivative()(np.log(x)), c=cmap(i/nrounds))
            x, y = x[len(x)//2:], y[len(y)//2:]
            m,b = np.polyfit(np.log(x), np.log(y), deg=1)
            # print(m)
            # axes[0].plot(data_amt, np.exp(b) * data_amt**m, linestyle='--', alpha=0.5, c=cmap(i/nrounds), lw=1)
    axes[0].axhline(baseline, c='black')
    axes[0].grid()
    axes[0].grid(which='minor', linestyle=':', linewidth='0.5', color='gray')
    axes[0].set_xlim(3e5, 5e7)
    axes[0].set_ylim(1e-3, 0.4)
    axes[1].set_ylim(0, 2.5)
    axes[1].grid()
    axes[1].grid(which='minor', linestyle=':', linewidth='0.5', color='gray')

fig, axes = plt.subplots(1, 2, figsize=(8, 4), sharex=True)
plot_perf(67, fig, axes, 7, 0.0588)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4), sharex=True)
fig.suptitle('resnet50')
plot_perf(73, fig, axes, 13, 0.0771, s=1.0)

fig, axes = plt.subplots(1, 2, figsize=(8, 4), sharex=True)
fig.suptitle('resnet101')
plot_perf(74, fig, axes, 13, 0.0771)

# fig, axes = plt.subplots(1, 2, figsize=(8, 4), sharex=True)
# fig.suptitle('resnet50 embedding d=128')
# plot_perf(75, fig, axes, 13, 0.0771)3

fig, axes = plt.subplots(1, 2, figsize=(8, 4), sharex=True)
fig.suptitle('resnet101 embedding d=32')
plot_perf(83, fig, axes, 13, 0.0771)
```
